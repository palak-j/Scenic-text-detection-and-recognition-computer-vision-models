EasyOCR is built with Python and Pytorch deep learning library, having a GPU could speed up the whole process of detection. The detection part is using the CRAFT algorithm and the Recognition model is CRNN.

EasyOCR has been integrated into Huggingface using Gradio.

Advantages: <br>
a)	Ready-to-use OCR with 80+ supported languages and all popular writing scripts including Latin, Chinese, Arabic, Devanagari, Cyrillic, etc. <br>
b)	Easy to use. <br>
c)	Can process multiple languages at the same time provided they are compatible with each other.

https://pypi.org/project/easyocr/

```python
import easyocr
import PIL
from PIL import ImageDraw
import time
import matplotlib.pyplot as plt
import os
```

```python
easyocr_reader = easyocr.Reader(['en'])           # this needs to run only once to load the model into memory
                                                  # can use 'ch_sim' in place of english 'en' language
```

```python
# draw boundation box

def draw_boxes(image_path, model):
    start_time = time.time()
    image = PIL.Image.open(image_path)
    res = model.readtext(image)
    
    draw = ImageDraw.Draw(image)
    for bound in res:
        p0,p1,p2,p3 = bound[0]
        draw.line([*p0, *p1, *p2, *p3, *p0], fill='red', width=4)
        
    end_time = time.time()
    
    return (image, res)
```

```python
# Test with sample images 

output = draw_boxes('images/img1.jpg', easyocr_reader)
plt.figure(figsize=(10,10))
plt.axis("off")
plt.imshow(output[0])
plt.show()

print(output[1])
```
